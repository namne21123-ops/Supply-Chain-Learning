```python
import pandas as pd

df = pd.read_csv("/Superstore.csv")

print(df.head())
```

```python
print(df.info())

print(df.describe())
```

```python
defective_order = df[df['Postal Code'].isnull()]

print("--- THE LIST OF 11 FUCKING CRIMINALS ---")
print(defective_order)

df['Postal Code'].fillna(5401, inplace=True)

print(df.info())
```

```python
top_customers = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(5)

print("--- üëë Top 5 VIP Customers ---")
print(top_customers)

print("\n-----------------------\n")

top_cities = df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(5)

print("--- üèôÔ∏è Top 5 Best Cities ---")
print(top_cities)
```

```python
top_cities.plot(kind='bar', figsize=(10,5), color='green', title="üèÜ TOP 5 CITIES BY SALES" )

```

```python
industries = df.groupby('Category')['Sales'].sum().sort_values(ascending=False).head(5)

print("--- Sales by INDUSTRIES ---")
print("industries")

industries.plot(kind='pie', figsize=(7,7), autopct='%1.1f%%', title="REVENUE STRUCTURE", ylabel="")
```

```python
# Time series (Yearly Sales)
df['Order Date'] = pd.to_datetime(df['Order Date'], format="%d/%m/%Y")

df['Year'] = df['Order Date'].dt.year

yearly_sales = df.groupby('Year')['Sales'].sum()

print("--- Yearly Sales ---")
print(yearly_sales)

yearly_sales.plot(kind='line', marker='o', figsize=(10,5), title='üìà SALES TREND (2014 - 2017)', grid=True, color='red')
```
