```python
import pandas as pd
from sklearn.linear_model import LinearRegression


data = {'size_sqft': [1000, 2000, 3000, 4000, 5000],
    'price': [200000, 400000, 600000, 800000, 1000000]}

df = pd.DataFrame(data)

x = df[['size_sqft']]
y = df['price']

model = LinearRegression()

model.fit(x, y)

print("Training Complete. The AI has learned the pattern.")

prediction = model.predict([[8500]])
# Note the 'f' before the quote "
print(f"Predicted Price for 8500 sq ft: ${prediction[0]:,.2f}")
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

data = {'size_sqft': [1000, 2000, 3000, 4000, 5000],
    'bedrooms':  [1,    2,    3,    4,    5],    # More rooms
    'age':       [10,   10,   10,   10,   10],   # Same age for now to keep it simple
    'price':     [200000, 420000, 640000, 860000, 1080000]
}

df = pd.DataFrame(data)

x = df[['size_sqft', 'bedrooms', 'age']]
y = df['price']

model = LinearRegression()
model.fit(x, y)
print("Multivariate Training Complete.")

prediction = model.predict([[8000, 8, 50]])

print(f"Predicted Price: ${prediction[0]:,.2f}")

import pandas as pd
from sklearn.linear_model import LinearRegression

data = {'size_sqft': [1000, 2000, 3000, 4000, 5000],
    'bedrooms':  [1,    2,    3,    4,    5],
    # Now we have variety! Some new (0), some old (20, 30)
    'age':       [0,    10,   20,   30,   40],
    # Prices are a bit lower for the older ones now
    'price':     [220000, 400000, 580000, 760000, 940000]}

df = pd.DataFrame(data)

x = df[['size_sqft', 'bedrooms', 'age']]
y = df['price']

model = LinearRegression()
model.fit(x, y)
print("Model Retrained with better Data.")

price_new = model.predict([[8000, 8, 0]])

print_old = model.predict([[8000, 8, 50]])

print(f"Price of NEW house: ${price_new[0]:,.2f}")
print(f"Price of OLD house: ${print_old[0]:,.2f}")
print(f"Value Lost due to age: ${price_new[0] - print_old[0]:,.2f}")
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

data = {'size_sqft': [1000, 1000, 2000, 2000, 3000, 3000],
    # Age varies independently of size
    # Small New, Small Old, Med New, Med Old, etc.
    'age':       [0,    50,   0,    50,   0,    50],
    # Prices: Notice how the OLD ones (50) are cheaper than the NEW ones (0) of the same size
    'price':     [200000, 150000, 400000, 350000, 600000, 550000]
}

df = pd.DataFrame(data)

model = LinearRegression()
model.fit(df[['size_sqft', 'age']], df['price'])

price_new = model.predict([[3000, 0]])

price_old = model.predict([[3000, 50]])

print(f"Price New: ${price_new[0]:,.2f}")
print(f"Price Old: ${price_old[0]:,.2f}")

diff = price_new[0] - price_old[0]
print(f"Value LOST due to age: ${diff:,.2f}")
```
